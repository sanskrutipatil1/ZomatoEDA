```python
country_names=final_df.Country.value_counts().index
print(country_names)
country_val=final_df.Country.value_counts().values
country_val
```

```python
# pie chart for top 3 countries
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
ratings
```

## observation

1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 4.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 1.8 to 2.4----> Poor
7. when Rating is 0 the customers has not rated.


```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (13, 7)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette=['pink','red','orange','yellow','green','green'])
```

## observation 
1. count for non rated is very high
2.rating is mostly between 2.5 to 4.0
